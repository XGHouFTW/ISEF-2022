```python
# This script will take in a large numerical dataset, and transform every column
# So that it is represented as a z-score (value - mean)/STD

import pandas as pd
import numpy as np
import sys
from google.colab import drive

drive.mount("/content/MyDrive/", force_remount=True)
```

```python
def normalize_column(column: pd.Series)-> pd.Series:
  STD = np.std(column)
  mean = np.mean(column)
  column = column.apply(lambda x: (x-mean)/STD)
  return column
```

```python
def main(data: pd.DataFrame):
  columns = list(data.columns)
  for column in columns[2:]:
    data[column] = normalize_column(data[column])
```

```python
if __name__=="__main__":
  filepath = './MyDrive/MyDrive/ISEF/Datasets/MergedPriceData.csv'
  df = pd.read_csv(filepath)
  main(df)
  display(df)
  # df.to_csv(filepath, index=False)
```
